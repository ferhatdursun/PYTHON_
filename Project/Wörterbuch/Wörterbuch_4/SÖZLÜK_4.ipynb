```python
#Sözlük

TrDE ={
    "Monate":
         {"Ocak":"Januar", "Subat":"Februar", "Mart":"März", "Nisan":"Äpril", "Mayis":"Mai", "Haziran":"Juni", "Temmuz":"Juli", "Agustos":"August","Eylül":"September","Ekim":"Oktober","Kasim":"November","Aralik":"Dezember"},
    "Zahlen":
         {"Bir":"Eins", "Iki":"Zwei", "Üc":"Drei", "Dört":"Vier", "Bes":"Fünf", "Alti":"Sechs", "Yedi":"sieben", "Sekiz":"Acht", "Dokuz":"Nine"},
    "Tiere":
         {"Kurt":"Wolf", "Aslan":"Löwe", "Sirtlan":"Hyäne", "Cakal":"Schakal","Köpek":"Hund"},
    "Jahresjeiten":
         {"Ilkbahar":"Frühling", "Yaz":"Sommer", "Sonbahar":"Herbst", "Winter":"Kis"},
    "Family":
         {"Baba":"Father", "Anne":"Mutter", "Erkek Kardes":"Bruder", "Kiz Kardes":"Schwester"},
}

def menu():
     print("[1]. Kelime Arama")
     print("[2]. Kelime Silme")
     print("[3]. Listeleme")
     print("[4]. Kelime Ekleme")
     print("[5]. Cikis")
     

     while True:
          giris = input("Lütfen menüden yapmak istediginiz islemin numarasini buraya yaziniz!:\n 1. Kelime arama\n2. Silme\n3.Listeleme\n4.Ekleme\n5.Cikis\n")

          if giris == "1": # 100% HAZIR.
               
               arama = input("Lütfen hangi sözlük alt basliginda arama yapmak istediginizi giriniz: Monate,Zahlen, Tiere, Jahresjeiten, Family ").title().strip()
               kelime= input("Lütfen arama yapmak istediginiz sözlük alt basliginda ki kelimeyi giriniz:").title().strip()
               for x, y in (TrDE[arama].items()):
                    if y == kelime:
                         x = (TrDE[arama].get(kelime,"Bu kelime bulunmamaktadir."))
                         print(x)
                         with open("/Users/ferhatdursun/Desktop/Clarusway/ITF/Python/Hausaufgabe/1_SÖZLÜKLER/Sözlük_4_Son_ödev/Kelime_Arama.txt", "w") as f:
                              f.writelines("Aradiginiz kelime: " + kelime + "\n" +"Aradiginiz kelimenin tercümesi: "+ x + "\n")

                    else:
                         if x == kelime:
                              y = (TrDE[arama].get(kelime,"Bu kelime bulunmamaktadir."))
                              print(y)
                              with open("/Users/ferhatdursun/Desktop/Clarusway/ITF/Python/Hausaufgabe/1_SÖZLÜKLER/Sözlük_4_Son_ödev/Kelime_Arama.txt", "w") as f:
                                   f.writelines("Aradiginiz kelime: " + kelime + "\n" +"Aradiginiz kelimenin tercümesi: "+ y + "\n")
                      
                               
          if giris == "2": # 100% HAZIR.
               silme = input("Lütfen silmek istediginiz kelimenin sözlük alt basligini giriniz: Monate, Zahlen, Tiere, Jahresjeiten, Family").title().strip()
               kelime = input("Lütfen silmek istediginiz kelimeyi giriniz: ").title().strip()
               a = (TrDE[silme].pop(kelime, "Bu kelime sözlükte bulunmamaktadir."))
               print("Silmek istediginiz kelime: "+ kelime + "\n" + "Silmek istediginiz kelimenin tercümesi: " + a)
               
               with open("/Users/ferhatdursun/Desktop/Clarusway/ITF/Python/Hausaufgabe/1_SÖZLÜKLER/Sözlük_4_Son_ödev/Silinenler.txt", "w") as sl:
                    for x,y in (TrDE[silme].items()):
                         fd = (x, ":", y,",","\t")
                         sl.writelines(fd)

                    sl.writelines("\nSilmek istediginiz kelime: " + kelime +"\n"+"Silmek istediginiz kelimenin tercümesi: "+ a)
                    


          if giris == "3": # 100% HAZIR.
               
               listele = input("Lütfen listelemek istediginiz sözlük alt basligini giriniz: Monate, Zahlen, Tiere, Jahresjeiten, Family:").title().strip()
               liste = (TrDE[listele])
               print(liste)
               with open("/Users/ferhatdursun/Desktop/Clarusway/ITF/Python/Hausaufgabe/1_SÖZLÜKLER/Sözlük_4_Son_ödev/listenenler.txt", "w") as ls:
                    for x, y in liste.items():
                         a = (x,":",y,"\t\t")
                         ls.writelines(a)
                    ls.writelines("\nListelemek istegidiniz sözlük alt basligi: " + listele)


          if giris == "4": #100% HAZIR. 
               ekleme = input("Lütfen eklemek istediginiz kelimenin sözlük alt basligini giriniz: Monate,Zahlen, Tiere, Jahresjeiten, Family").title().strip()
               kelime1 = input("Lütfen eklemek istediginiz birinci kelimeyi giriniz:").title().strip()
               kelime2 = input("Lütfen eklemek istediginiz ikinci kelimeyi giriniz: ").title().strip()

               (TrDE[ekleme].update({kelime1:kelime2}))
               print(TrDE[ekleme])
               
               with open("/Users/ferhatdursun/Desktop/Clarusway/ITF/Python/Hausaufgabe/1_SÖZLÜKLER/Sözlük_4_Son_ödev/Eklenenler.txt", "w") as ek:
                    
                    for x,y in (TrDE[ekleme].items()):
                         f = (x,":",y,"\t\t")                                                 
                         ek.writelines(f)
                    ek.writelines("\nEklemek istediginiz kelimeler; "+ x + ":" + y)
                    ek.writelines("\nKelime eklemek istediginiz sözlük alt basligi; "+ ekleme + "\n")
                    
          else:

               print("Cikis yapiliyor.")


          break

menu()

```
